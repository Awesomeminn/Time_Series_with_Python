```python
import numpy as np 

np.random.seed(42)

steps = np.random.standard_normal(1000)
steps[0] = 0

random_walk = np.cumsum(steps)
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()

ax.plot(random_walk)
ax.set_xlabel('TimeSteps')
ax.set_ylabel('Value')
```

```python
# 생성된 Random Walk을 이용하여 ADF 테스트 해보기
from statsmodels.tsa.stattools import adfuller

ADF_result = adfuller(random_walk)
print(f'ADF Statistics: {ADF_result[0]}')
print(f'p-value: {ADF_result[1]}') ## p-value가 0.05보다 크므로 비정상성이라는 귀무가설을 기각할 수 없다.
```

```python
print(ADF_result)
```

```python
# ACF 도식화
from statsmodels.graphics.tsaplots import plot_acf

plot_acf(random_walk, lags = 100)
```

```python
# Random Walk 1차 차분 후 ADF 검정
diff_random_walk = np.diff(random_walk, n = 1)
```

```python
fig, ax = plt.subplots()

ax.plot(diff_random_walk)
ax.set_xlabel('TimeSteps')
ax.set_ylabel('Value')
```

```python
ADF_result = adfuller(diff_random_walk)

print(f'ADF Statistics: {ADF_result[0]}')
print(f'p-value: {ADF_result[1]}') ## 귀무가설을 기각하게 되면서 정상성을 만족한다고 할 수 있다.
```

```python
# 1차 차분한 랜덤워크의 ACF 도식화
plot_acf(diff_random_walk, lags = 20)
```

```python
# 차분한 종가를 가지고 다시 ADF 검정 
GOOGL_diff_ADF_result = adfuller(diff_close)

print(f'ADF Statistics: {GOOGL_diff_ADF_result[0]}')
print(f'p-value: {GOOGL_diff_ADF_result[1]}') ## 1계 차분된 종가는 p-value가 0.05보다 훨씬 작아 비정상성이라는 귀무가설을 기각
```

```python
# ACF 함수를 도식화하여 자기상관관계가 있는지 확인
from statsmodels.graphics.tsaplots import plot_acf
plot_acf(diff_close, lags = 20)
```

# 확률보행 예측하기 

```python
# 긴 기간 예측하기
import pandas as pd

df = pd.DataFrame({'value': random_walk})
```

```python
fig, ax = plt.subplots()

ax.plot(random_walk)
ax.set_xlabel('TimeSteps')
ax.set_ylabel('Value')
```

```python
## 훈련 데이터와 테스트 데이터 나누기
train = df[:800]
test = df[800:]
```

```python
## 베이스라인1: 과거값들의 평균
mean = np.mean(train.value)

test.loc[:, 'pred_mean'] = mean

test.head()
```

```python
## 베이스라인2: 훈련 데이터셋에서 가장 마지막으로 측정된 값
last_value = train.iloc[-1].value
test.loc[:, 'pred_last'] = last_value
test.head()
```

```python
## 베이스라인3: drift method 
deltaX = 800 - 1
deltaY = last_value

drift = deltaY / deltaX
print(f'drift: {drift}')

x_vals = np.arange(801, 1001, 1)

pred_drift = drift * x_vals

test.loc[:, 'pred_drift'] = pred_drift

test.head()
```

```python
## 세 가지 방법에 의한 예측값을 테스트 데이터의 실젯값과 비교해볼 수 있도록한 시각화 실시
fig, ax = plt.subplots()

ax.plot(train.value, 'b-')
ax.plot(test['value'], 'b-')

ax.plot(test['pred_mean'], 'r-.', label = "Mean")
ax.plot(test['pred_last'], 'g--', label = 'Last_value')
ax.plot(test['pred_drift'], 'k:', label = 'Drift')

ax.axvspan(800, 1000, color = '#808080', alpha = 0.2)
ax.legend(loc = 2)

ax.set_xlabel('Timesteps')
ax.set_ylabel('Value')

plt.tight_layout()
```

```python
## 세 기법에 대한 MSE 계산
from sklearn.metrics import mean_squared_error

mse_mean = mean_squared_error(test['value'], test['pred_mean'])
mse_last = mean_squared_error(test['value'], test['pred_last'])
mse_drift = mean_squared_error(test['value'], test['pred_drift'])

print(mse_mean, mse_last, mse_drift)
```

```python
## MSE 시각화
# 변수 이름과 데이터를 리스트로 정리
variables = ["Mean", "Last value", "Drift"]
values = [mse_mean, mse_last, mse_drift]

# 그래프 생성
x = np.arange(len(variables))  # x축 위치
bar_width = 0.6  # 막대 너비

plt.figure(figsize=(8, 6))
bars = plt.bar(x, values, width=bar_width, color=["blue", "green", "red"], alpha=0.7)

# 막대 위에 값 표시
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height, f"{height:.1f}", 
             ha='center', va='bottom', fontsize=10)

# 그래프 설정
plt.xticks(x, variables, rotation=45)  # x축 레이블과 각도
plt.xlabel("Method")
plt.ylabel("MSE")
plt.title("Comparison of MSE for Different Methods")
plt.tight_layout()

# 그래프 표시
plt.show()
```

```python
# 다음 시간 단계 예측하기 
## 한 시점의 관측값을 바로 다음 시점의 예측값으로 사용한다.
df_shift = df.shift(periods = 1)
df_shift.head()
```

```python
## 예측 도식화

fig, ax = plt.subplots()

ax.plot(df, 'b-', label = 'actual')
ax.plot(df_shift, 'r-.', label = 'forecast')

ax.legend(loc=2)

ax.set_xlabel('Timesteps')
ax.set_ylabel('Value')

plt.tight_layout()
```

```python
mse_one_step = mean_squared_error(test['value'], df_shift[800:])
print(mse_one_step)
```
